```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate
import math
```

```python
ds = 0.01
g = 9.8
c = 3
v = 20
# set up array of times
# initialize array S, I and R
n = 5000
T = np.zeros(n)
X = np.zeros(n)
Y = np.zeros(n)
# initialize S[0], I[0], R[0]
T[0] = 0
X[0] = 0
Y[0] = 0
# set up for loop
for j in range(0, n-1):
  T_prime = 3/(40.82-(2*Y[j]))
  T[j+1]=T[j]+T_prime*ds
  X_prime = np.cos(T[j])
  Y_prime = np.sin(T[j])
# do Euler step
  X[j+1]=X[j]+X_prime*ds
  Y[j+1]=Y[j]+Y_prime*ds
plt.plot(X,Y)
```
